```python
# {i}_iter_num
iter_num_cols_list = [f"{i}_iter_num" for i in range(100)]
```

```python
rilo_2[iter_num_cols_list].apply(pd.value_counts).fillna(0).sum(axis=1).plot(kind="bar")
```

```python
rilo_3[iter_num_cols_list].apply(pd.value_counts).fillna(0).sum(axis=1).plot(kind="bar")
```

```python
rilo_4[iter_num_cols_list].apply(pd.value_counts).fillna(0).sum(axis=1).reset_index()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# データセットのリストとタイトル
datasets = [rilo_2, rilo_3, rilo_4]
titles = ["k=2", "k=3", "k=4"]

# 各データに対してバーグラフを作成
for i, (df, title) in enumerate(zip(datasets, titles)):
    tmp_df = df[iter_num_cols_list].apply(pd.value_counts).fillna(0).sum(axis=1).reset_index()
    axes[i].bar(
        tmp_df["index"], tmp_df[0]
    )
    axes[i].set_title(title)  # タイトル設定
    axes[i].set_xlabel("Value")  # X軸ラベル
    axes[i].set_ylabel("Count")  # Y軸ラベル
    axes[i].set_xlim(1, 7)

# レイアウト調整
plt.tight_layout()
plt.show()
```
